#Write a Python program to sum all the items in a list.

```python
def list_sum(arr):
    sum_num=0
    for num in arr:
        sum_num += num
    return sum_num
```

```python
list_sum([5,4,3])
```

Write a Python program to count the number of strings where the string length is 2 or
more and the first and last character are same from a given list of strings.

```python
def checker(arr):
    return len(list(filter(lambda x : len(x)>1 and x[0]==x[-1],arr)))
```

```python
checker(['abc', 'xyz', 'aba', '1221','bhgsskknb','aa'])
```

#Write a Python program to remove duplicates from a list.

```python
def remove_dup(arr):
    return list(set(arr))
```

```python
a = remove_dup([3,4,4,5,5,7,7,9])
```
