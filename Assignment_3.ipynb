### Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def myreduce(func,items):
    try:
        if(len(items)==0):
            print("TypeError: reduce() of empty sequence with no initial value")
        elif(len(items)==1):
            return items[0]
        else:
            a,b = items[0],items[1]
            result = func(a,b)
            for a in range(2,len(items)):
                result = func(result,items[a])
    except :
        print("Something went wrong")
    return result
```

```python
myreduce(lambda x,y:x*y,[1,2,3,4])
```

### Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def myfilter(func,items):
    result = []
    for a in items:
        if(func(a)):
            result.append(a)
    return result
```

```python
list(myfilter(lambda x:x>5,[1,2,3,4,5,6,30,40,20]))
```

###  Implement List comprehensions to produce the following lists.
###  Write List comprehensions to produce the following Lists
### ['x','xx','xxx','xxxx','y','yy','yyy','yyyy','z','zz','zzz','zzzz']
### ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']
### [[2], [3], [4], [3], [4], [5], [4], [5], [6]] [[2, 3, 4, 5], [3, 4, 5, 6],[4, 5, 6, 7], [5, 6, 7, 8]]
### [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]

```python
[a*b for a in ['x','y','z'] for b in range(1,5)]
```

```python
[a*b for a in range(1,5) for b in ['x','y','z']]
```

```python
[[b] for a in range(2,5) for b in range(a,a+3)]
```

```python
[list(range(a,a+4)) for a in range(2,6)]
```

```python
[(b,a) for a in range(1,4) for b in range(1,4)]
```
