#Create a list of length 3 and Create a list of length 4, For each element in the first list,

```python
first = [1,2,3]
second = [4,3,2,1]
for val in zip(first,second):
    print("first element is {} and corresponding element in second list is {}".format(val[0],val[1]))
```
